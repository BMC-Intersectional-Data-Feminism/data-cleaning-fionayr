```python
import pandas as pd 
import numpy as np
```

```python
df_example = pd.read_csv('audible_uncleaned.csv') # read in csv file
df_example
```

```python
## AUTHOR 
df_example['author'] = df_example['author'].str.strip('Writtenby:')
df_example
```

```python
import matplotlib.pyplot as plt #import package for visualization

valueCounts = df_example['author'].head(100).value_counts() # counts first 100 rows that author appears

valueCounts.plot(kind='bar') 
```

```python
## NARRATOR
df_example['narrator'] = df_example['narrator'].str.strip('Narratedby:')
df_example
```

```python
## TIME
def convertTime(timeString):
    totalMins = 0
    splitString = timeString.split(' and ') #split on and statement 
    for item in splitString:
        if 'hrs' in item:
            hours = int(item.split(' ')[0])
            totalMins += (60*hours)
        elif 'mins' in item:
            mins = int(item.split(' ')[0])
            totalMins += mins
    return totalMins

convertTime("2 hrs and 10 mins")
# df_example['time'] = 
# df_example
```

```python
df_example['time'] = df_example['time'].apply(convertTime) # how to apply a function + link
df_example
```

```python
## RELEASE DATE

df_example['releasedate'] = pd.to_datetime(df_example["releasedate"], infer_datetime_format=True, format='mixed').dt.strftime('%Y-%m-%d')
df_example
```

```python
## STARS

notRated = df_example[df_example['stars'] == 'Not rated yet'] # select the rows where nothing is rated
rated = df_example[df_example['stars'] != 'Not rated yet'] # select all the rows where there is a rating

rated
```

```python
rated[['stars', 'numRatings']] = rated['stars'].str.split('stars', expand = True)
rated
```

```python
rated['stars'] = rated['stars'].str.split(' out of ').str[0]
rated
```

```python
df_cleaned = pd.concat[rated, notRated]
df_cleaned
```
